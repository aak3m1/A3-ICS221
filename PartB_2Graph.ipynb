```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.DiGraph() #creating a graph object

intersections = {0: (0, 1), 1: (1, 2), 2: (1, 0)} #defining the nodes with their positions and labels
houses = {4: (2, 2), 5: (2, 0)}
all_nodes = {**intersections, **houses}

for node_id, pos in all_nodes.items(): #adding intersections and houses to the graph
    G.add_node(node_id, pos=pos)

edges = [(0, 1, 5), (1, 2, 3), (0, 2, 7), (1, 4, 0.1), (2, 5, 0.1)] #definig the edges with travel times

for start, end, weight in edges: #adding the edges to the graph
    G.add_edge(start, end, weight=weight)

pos = nx.get_node_attributes(G, 'pos') # the position the nodes using their defined positions


nx.draw_networkx_nodes(G, pos, nodelist=intersections.keys(), node_color='skyblue', node_size=500)
nx.draw_networkx_nodes(G, pos, nodelist=houses.keys(), node_color='green', node_shape='s', node_size=400)
nx.draw_networkx_edges(G, pos, edgelist=edges, arrowstyle='-|>', arrowsize=15)
nx.draw_networkx_labels(G, pos, font_size=10, font_color= "white")

edge_labels = {(start, end): G[start][end]['weight'] for start, end in G.edges()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12)

plt.title('Package Delivery Routes from Warehouse')
plt.axis('off')
plt.show()
```
