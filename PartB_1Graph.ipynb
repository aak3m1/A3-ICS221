```python
import matplotlib.pyplot as plt
import networkx as nx

#itializing a directed graph to represent the traffic network where nodes are intersections and edges are roads
G = nx.DiGraph()

#adding a nodes to the graph; each node represents an intersection in the city
intersections = [0, 1, 2] #example intersection IDs
for intersection in intersections:
    G.add_node(intersection, label=f'Intersection {intersection}') #label each node with its intersection ID

# defining the roads connecting intersections with travel times
# for each tuple in roads list represents a road with start node, end node, and weight (travel time in minutes)
roads = [(0, 1, 5), (1, 2, 3), (0, 2, 7)]
for start, end, length in roads:
    G.add_edge(start, end, weight=length, label=f'{length} min') # adding the edges with labels showing travel times

#arranging the nodes using a circular layout for better visual clarity and to emphasize network structure
pos = nx.circular_layout(G)

#drawing the nodes with specified options for better aesthetics
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=500) #larger, sky blue nodes for visibility

#drawinf edges with labels
edges = nx.draw_networkx_edges(G, pos, edgelist=roads, arrowstyle='-|>', arrowsize=15, style='dashed')
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

#drawing a node labels with larger fonts for better readability
nx.draw_networkx_labels(G, pos, font_size=10, font_color='white')

plt.title('City Traffic Graph with Optimized Routes')
plt.axis('off')  # Turn off the axis
plt.show()  # Display the plot
```
