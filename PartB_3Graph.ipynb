```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph() #graph object
G.add_nodes_from(range(7))  #nodes to the graph (intersections and houses)  from 0 to 6

edges = [
    (0, 1, 4), (0, 2, 2), (1, 2, 3), (1, 3, 2),
    (2, 3, 6), (2, 4, 1), (3, 4, 1), (3, 5, 5),
    (4, 5, 2), (5, 6, 3)
] #(the roads between the nodes)

G.add_weighted_edges_from(edges)

shortest_path = nx.shortest_path(G, source=0, target=4, weight='weight') #shortest path from point A (0) to point B (4)
pos = nx.spring_layout(G) #position map for nodes for better visualization
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=700) #nodes
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='gray', width=2) #edges
nx.draw_networkx_edges(G, pos, edgelist=list(zip(shortest_path, shortest_path[1:])), edge_color='blue', width=3) #shortest path
nx.draw_networkx_labels(G, pos, font_size=14, font_family='sans-serif', font_color='black') #node labels

#Edge weight labels
edge_weights = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_weights)

#plot title
plt.title('Shortest Path from Point A to Point B', size=12)
plt.axis('off')
plt.tight_layout()
plt.show()
```
